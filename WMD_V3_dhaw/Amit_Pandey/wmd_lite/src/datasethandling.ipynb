```python
import os
import numpy as np
import json
from collections import Counter, defaultdict
```

```python
## Store glove embeddings as dictionary for later use:

dim = [50,100,200,300]

for i in dim:
    embeddings_dict = defaultdict(lambda:np.zeros(i)) 
    # defaultdict to take care of OOV words.
    
    with open(f"../files/glove.6B.{i}d.txt",'r') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], "float32")
            embeddings_dict[word] = vector.tolist() ## to avoid json storage error as it cant serialise array
        
    embeddings_dict = dict(embeddings_dict)
    
    ## insteaf of dump in one line, dump each key value pari and then add new line
    ##https://stackoverflow.com/questions/16968198/loading-and-reading-json-file-is-very-slow-after-writing-in-python
    
    a_file = open(f"../files/glove{i}d_embdict.json", "w")
    json.dump(embeddings_dict, a_file)
    a_file.close()
    
    '''
    # a_file = open("data.json", "r")
    # output = a_file.read()
    # print(output)

    # a_file.close() '''
```
